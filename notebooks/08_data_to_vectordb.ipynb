```python
import numpy as np
import json
from cortex import CortexClient, DistanceMetric
import os
```

```python

embeddings = np.load("data/embeddings/merged_cancer_embeddings_all_sources.npy")
with open("data/records/merged_cancer_records_all_sources.json", "r") as f:
    records = json.load(f)
```

```python

with CortexClient("localhost:8000") as client:
    collection_name = "hetionet_final"
    
    if client.has_collection(collection_name):
        client.delete_collection(collection_name)
    
    client.create_collection(
        name=collection_name, 
        dimension=384, 
        distance_metric=DistanceMetric.COSINE
    )
    
    ids = list(range(len(records)))
    client.batch_upsert(
        collection_name,
        ids,
        embeddings.tolist(),
        records  
    )

print("Data successfully migrated to Actian!")
```
