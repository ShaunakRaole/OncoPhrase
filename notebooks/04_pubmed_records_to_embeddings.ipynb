```python
from sentence_transformers import SentenceTransformer
import json
import numpy as np

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)
```

```python
with open("data/records/pubmed_records_cancer_mvp.json", "r") as f:
    pubmed_records = json.load(f)

print("Loaded records:", len(pubmed_records))
```

```python
for r in pubmed_records:
    r["source"] = "pubmed"
# Again, making sure each record has search_text
pubmed_records[0].keys()
```

```python
texts = [r["search_text"] for r in pubmed_records]

pubmed_embeddings = model.encode(
    texts,
    convert_to_numpy=True,
    normalize_embeddings=True,   # same as Hetionet
    show_progress_bar=True,
    batch_size=32
)

print("Embeddings shape:", pubmed_embeddings.shape)
```

```python
np.save("data/embeddings/pubmed_embeddings_cancer_mvp.npy", pubmed_embeddings)
print("Saved pubmed_embeddings_cancer_mvp.npy")
```
