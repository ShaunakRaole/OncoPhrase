```python
import json
import numpy as np
from pathlib import Path
from sentence_transformers import SentenceTransformer

model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)
```

```python
records_path = Path("data/records/clinical_trials_records_cancer_mvp.json")
```

```python
# Load records
with open(records_path, "r") as f:
    clinical_trials_records = json.load(f)

print("Loaded records:", len(clinical_trials_records))
print("Example keys:", clinical_trials_records[0].keys())
print("Example text:", clinical_trials_records[0]["search_text"][:300])
```

```python
texts = [r["search_text"] for r in clinical_trials_records]

clinical_trials_embeddings = model.encode(
    texts,
    convert_to_numpy=True,
    normalize_embeddings=True,
    show_progress_bar=True,
    batch_size=32
)

print("Clinical trial embeddings shape:", clinical_trials_embeddings.shape)
print("dtype:", clinical_trials_embeddings.dtype)
```

```python
out_npy = Path("data/embeddings/clinical_trials_embeddings_cancer_mvp.npy")
np.save(out_npy, clinical_trials_embeddings)

print("Saved:", out_npy)
```
